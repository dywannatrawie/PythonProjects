```python
import pandas as pd
import numpy as np

# Load the Excel file
input_file = 'HunterLeads.xlsx'
output_file = 'MailDataBase.xlsx'
excel_data = pd.ExcelFile(input_file)

# Create an empty dataframe to store merged data
merged_data = pd.DataFrame()
```

```python
# Iterate through each sheet in the Excel file
for sheet_name in excel_data.sheet_names:
    # Read each sheet into a dataframe
    sheet_data = pd.read_excel(input_file, sheet_name=sheet_name)
    
    # Merge the current sheet with the existing merged data on "Email"
    if merged_data.empty:
        merged_data = sheet_data
    else:
        merged_data = pd.concat([merged_data, sheet_data], axis=0, ignore_index=True).drop_duplicates(subset='Email', keep='first')

# Handle duplicate columns by combining them
for column in merged_data.columns:
    if '_dup' in column:
        original_column = column.replace('_dup', '')
        if original_column in merged_data.columns:
            merged_data[original_column] = merged_data[original_column].fillna(merged_data[column])
            merged_data.drop(columns=[column], inplace=True)

# Write the merged data to a new Excel file
with pd.ExcelWriter(output_file) as writer:
    merged_data.to_excel(writer, index=False, sheet_name='MergedData')

print(f"Data merged successfully into {output_file}")
```
